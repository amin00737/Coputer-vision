```python
target = []
data = []
dir='/content/gdrive/MyDrive/DRCAP/Dataset/Cross validation dataset/Dataset'
Categories = ['1','3','4']

size=124
```

```python
for i in Categories:
  print("Category is:",i,"\tLabel encoded as:",Categories.index(i))
  target_class = Categories.index(i)
  path = os.path.join(dir,i)
  for img in os.listdir(path):
    img_array = cv2.imread(os.path.join(path,img))
    img_array = cv2.cvtColor(img_array, cv2.COLOR_BGR2RGB)
    img_resized =cv2.resize(img_array,(size,size))
    data.append(img_resized)
    target.append(target_class)

data = np.array(data)
target = np.array(target)

print('end')
```

```python
X_train,X_test,y_train,y_test = train_test_split(data,target,test_size=0.05)

X_train.shape
```

```python
X_train = X_train / 255.0
X_test =  X_test / 255.0
```

```python
custom_model = keras.models.load_model('/content/gdrive/MyDrive/DRCAP/Analyse/Classification/My Model/Apple_Amin1(0.9833).h5')


feature_extractor_layer = custom_model.get_layer('flatten_2')

VGG_Apple_Model1 = Model(inputs=custom_model.input, outputs=feature_extractor_layer.output)

A_extractor_feature=VGG_Apple_Model1.predict(X_train)
features_A_t = A_extractor_feature.reshape(A_extractor_feature.shape[0], -1)


```

```python
FeatureX_test = VGG_Apple_Model1.predict(X_test)
FeatureX_test1 = FeatureX_test.reshape(FeatureX_test.shape[0], -1)
```

```python
# features vector
features_vector = VGG_Apple_Model1.predict(X_train)  # Replace with your actual features

# custom Regressor
features_shape = features_vector.shape[1]  # Get the shape of the features vector

regressor = keras.Sequential([
    layers.Input(shape=(features_shape,)),
    layers.Dense(128, activation='relu'),
    layers.Dense(64, activation='relu'),
    layers.Dense(32, activation='relu'),
    layers.Dense(1)
])

# loss function and optimizer
regressor.compile(loss='mean_squared_error', optimizer='adam')

# summary of the model architecture
regressor.summary()
```

```python

regression_targets_train = y_train
regression_targets_test = y_test


regressor.fit(features_vector, regression_targets_train, epochs=50, batch_size=32, validation_split=0.2)


evaluation = regressor.evaluate(FeatureX_test1, regression_targets_test)


predictions = regressor.predict(FeatureX_test1)


mae = mean_absolute_error(regression_targets_test, predictions)


print("Mean Squared Error on Test Data (MSE):", evaluation)
print("Mean Absolute Error on Test Data (MAE):", mae)
```

```python
r2 = r2_score(regression_targets_test, predictions)
print("R-squared on Test Data (R2):", r2)
```
